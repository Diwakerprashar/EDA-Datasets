```python
#train test split
from sklearn.model_selection import train_test_split
```

```python
X = df[['feature1','feature2']].values
y = df['price'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# scale the data using minmaxScaler
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test =scaler.transform(X_test)
```

```python
## Starting with Keras for prediction model
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#help(Sequential)
```

#### Model - as a list of layers
#### Dense specifies Depth of neurons and units specifies no of neurons for that layer
##Dense(units=2),   layer 1 to have 2 neurons
#Dense(units=3),   layer 2 to have 3 neurons
#Dense(units=1)    layer 3 to have 1 neurons

model = Sequential([Dense(units=2),
                   Dense(units=3),
                   Dense(units=1)])


```python
### Model - adding in layers one by one

# this method is more useful as we can modify neurons at any stage and can remove neurons at any stage by commenting it.

model = Sequential()

model.add(Dense(2)) #no of neurons at layer 1
model.add(Dense(2)) #at layer 2
model.add(Dense(2)) # at layer3
```

```python
# adding the parameters to our model

model = Sequential()

model.add(Dense(2,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))

#final node donot have activation so don't add activation function
model.add(Dense(1))

model.compile(optimizer='rmsprop',loss='mse') #choosen based on type of prediction we are looking for, as here we are looking 
                                            # regression line to predict price so we are using mse
    
```

For a multi-class classification problem
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

For a binary classification problem
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])

For a mean squared error regression problem
model.compile(optimizer='rmsprop',
              loss='mse')

```python
model.fit(X_train,y_train,epochs=250) #epochs are the no of iterations that the model should do i.e model willdo learning 250 times here
# model has a parameter vervose which helps us to know on which iteration epochs our model is, if we set to False we will not see the below output
```
