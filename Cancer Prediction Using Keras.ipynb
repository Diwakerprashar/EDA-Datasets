# Train Test split

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
X = df.drop('benign_0__mal_1',axis=1)
y = df['benign_0__mal_1']
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=101)
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

## Tensorflow

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation,Dropout
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential()
```

```python
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))

# Binary Calssification
model.add(Dense(1,activation='sigmoid'))

#complie
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model.fit(X_train,y_train,epochs=600,validation_data=(X_test,y_test))
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
earlt_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
```

```python
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))

# Binary Calssification
model.add(Dense(1,activation='sigmoid'))

#complie
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model.fit(X_train,y_train,epochs=600,validation_data=(X_test,y_test),callbacks=earlt_stop)
```

```python
losss = pd.DataFrame(model.history.history)
losss.plot()
```

```python
model.add(Dense(30,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(15,activation='relu'))
model.add(Dropout(0.5))

# Binary Calssification
model.add(Dense(1,activation='sigmoid'))

#complie
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model.fit(X_train,y_train,epochs=600,validation_data=(X_test,y_test),callbacks=earlt_stop)
```

```python
lossss = pd.DataFrame(model.history.history)
lossss.plot()
```
