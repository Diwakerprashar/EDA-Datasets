### PCA for Dimention reduction and analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df.head()
```

```python
# the data is in 30 dimentions , we cannot plot and it's difficult to do analysis and ml predictions
df.shape
```

```python
# first we will scale the dataset

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
scaled_data
```

```python
# PCA

from sklearn.decomposition import PCA
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
# we have reduced 30 dimentions to 2 dimentions
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'],cmap='plasma')
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_,columns=cancer['feature_names'])
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma',)
```

```python
pca_3 = PCA(n_components=3)
```

```python
pca_3.fit(scaled_data)
```

```python
x_pca3 = pca_3.transform(scaled_data)
```

```python
x_pca3.shape
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111,projection='3d')
plt.scatter(x_pca3[:,0],x_pca3[:,1],x_pca3[:,2],c=cancer['target'],cmap='rainbow')
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
#plt.zlabel('Third Principal Component')
```

```python
cdict = {0:'m',1:'c'}
label = {0:'Malignant',1:'Benign'}
y =cancer['target']
```

```python
fig = plt.figure(figsize=(14,9))
ax = fig.add_subplot(111, projection='3d')
 
for l in np.unique(y):
 ix=np.where(y==l)
 ax.scatter(x_pca3[:,0],x_pca3[:,1],x_pca3[:,2],c=cdict[l],s=60,label=label[l])
 
ax.set_xlabel("PC1",fontsize=12)
ax.set_ylabel("PC2",fontsize=12)
ax.set_zlabel("PC3",fontsize=12)
 
ax.view_init(30, 125)
ax.legend()
plt.title("3D PCA plot")
plt.show()
```
