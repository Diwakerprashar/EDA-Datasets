## Scaling and training data

```python
X=df.drop('price',axis=1)
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
mmscaler = MinMaxScaler()
```

```python
mmscaler.fit(X_train)
```

```python
X_train=mmscaler.transform(X_train)
X_test=mmscaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation,Dense
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential()

model.add(Dense(units=19,activation='relu'))
model.add(Dense(units=19,activation='relu'))
model.add(Dense(units=19,activation='relu'))
model.add(Dense(units=19,activation='relu'))

model.add(Dense(units=1))

model.compile(optimizer='adam',loss='mse')
```

## Training Model

```python
model.fit(x=X_train,y=y_train.values,validation_data=(X_test,y_test.values),batch_size=128,epochs=400)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses #adding that validation data in training helps us to track the loss and we get a dataframe for loss in training data and validation data
```

```python
losses.plot() #adding that validation data in training helps us to track the loss and we can visualise through plot
```

```python
#let update the epochs and see how loss varies
```

```python
model.fit(x=X_train,y=y_train.values,validation_data=(X_test,y_test.values),batch_size=128,epochs=600)
```

```python
losses2 = pd.DataFrame(model.history.history)
```

### Model starts overfitting

```python
losses2.plot()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
predictions
```

```python
mean_squared_error(y_test,predictions)
```

```python
np.sqrt(mean_squared_error(y_test,predictions)) #root mse
```

```python
mean_absolute_error(y_test,predictions)
```

```python
explained_variance_score(y_test,predictions) #ranges between 0 and 1  and 1 being the best and here we have 81% accuracy
```

```python
plt.figure(figsize=(12,5))
plt.scatter(y_test,predictions,marker='*',c='r')
plt.plot(y_test,y_test,'g')
```
